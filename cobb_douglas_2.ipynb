```julia
@gif for i in range(0, stop = 1, length = 100)
    α = i;
    
    x= [1.0:0.3:10;];
    y= [1.0:0.3:10;];
    
    k(x,y) = x^(α) * y^(1-α);
    
    plot(x, y, k, st=:surface);
    end
```
